```python
dataset_name = 'Handwriting'
dataset_folder = './datasets/Handwriting'
```

```python
from fedot_ind.tools.loader import DataLoader

loader = DataLoader(dataset_name=dataset_name, folder=dataset_name)
train_data, test_data = loader.load_data()

assert isinstance(train_data, tuple) and isinstance(test_data, tuple)
```

```python
from fedot_ind.api.main import FedotIndustrial

api_config = dict(problem='classification',
                  metric='f1',
                  timeout=1,
                  pop_size=10,
                  with_tuning=False,
                  n_jobs=2,
                  logging_level=20)
fedot_ind = FedotIndustrial(**api_config)
fedot_ind.fit(train_data)

labels = fedot_ind.predict(test_data)
probs = fedot_ind.predict_proba(test_data)
metrics = fedot_ind.get_metrics(target=test_data[1],
                                rounding_order=3,
                                metric_names=('f1', 'accuracy', 'precision', 'roc_auc'))
result_dict = dict(industrial_model=fedot_ind,
                   labels=labels,
                   metrics=metrics)
result_dict
```
